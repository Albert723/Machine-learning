### Matplotlib基础

```python
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
import numpy as np
x = np.linspace(0,10,100)
```

```python
x
```

```python
y = np.sin(x)
```

```python
y
```

```python
plt.plot(x,y)
```

```python
plt.show()
```

```python
cosy = np.cos(x)
siny = np.sin(x)
plt.plot(x,siny)
plt.plot(x,cosy,color = "red")
plt.show()
```

```python
cosy = np.cos(x)
siny = np.sin(x)
plt.plot(x,siny)
plt.plot(x,cosy,color = "red",linestyle="--")
plt.show()
```

```python
cosy = np.cos(x)
siny = np.sin(x)
plt.plot(x,siny)
plt.plot(x,cosy,color = "red",linestyle="--")
plt.xlim(-5,15)
plt.show()
```

```python
cosy = np.cos(x)
siny = np.sin(x)
plt.plot(x,siny)
plt.plot(x,cosy,color = "red",linestyle="--")
plt.xlim(-5,15)
plt.ylim(0,1.5)
plt.show()
```

```python
cosy = np.cos(x)
siny = np.sin(x)
plt.plot(x,siny)
plt.plot(x,cosy,color = "red",linestyle="--")
plt.axis([-1,11,-2,2])
plt.show()
```

```python
cosy = np.cos(x)
siny = np.sin(x)
plt.plot(x,siny)
plt.plot(x,cosy,color = "red",linestyle="--")
plt.xlabel("x_axis")
plt.ylabel("y_value")
plt.show()
```

```python
cosy = np.cos(x)
siny = np.sin(x)
plt.plot(x,siny,label="sin(x)")
plt.plot(x,cosy,color = "red",linestyle="--",label="cos(x)")
plt.xlabel("x_axis")
plt.ylabel("y_value")
plt.legend()
plt.show()
```

```python
plt.plot(x,siny,label="sin(x)")
plt.plot(x,cosy,color = "red",linestyle="--",label="cos(x)")
plt.xlabel("x_axis")
plt.ylabel("y_value")
plt.legend()
plt.title("Welcome to ML")
plt.show()
```

### Scatter Plot(散点图）

```python
plt.scatter(x,y)
```

```python
plt.scatter(x,siny)
plt.show()
```

```python
plt.scatter(x,siny)
plt.scatter(x,cosy,color="red")
plt.show()
```

```python
g = np.random.normal(0,1,100)
h = np.random.normal(0,1,100)
plt.scatter(g,h)
plt.show()
```

```python
g = np.random.normal(0,1,10000)
h = np.random.normal(0,1,10000)
plt.scatter(g,h,alpha=0.1)
plt.show()
```
